In the following code syncronization is performed beacuse the central camera has different frame rate respect to the top right camera. The first frames are deleted.

```python
import os




def syncronization(directory_path, files_to_remove):
    file_list = sorted(os.listdir(directory_path))


    for file_name in file_list[:files_to_remove]:
        os.remove(os.path.join(directory_path, file_name))


    for index, file_name in enumerate(sorted(os.listdir(directory_path)), start=1):
        os.rename(
            os.path.join(directory_path, file_name),
            os.path.join(directory_path, f"{index:06d}.jpg")
        )
syncronization("/content/task_data/images/01", 260)
syncronization("/content/task_data/images/02", 171)


```

```python
import os
import json
import shutil

# Define the path to your main directory
source_directory = '/content/task_data/annotsleft'

destination_directory = source_directory.replace("annotsleft", "annots")

# Get a list of subdirectories within the main directory
subdirectories = [folder.path for folder in os.scandir(source_directory)]

# Iterate through each subdirectory
for subdirectory_path in subdirectories:

    # Retrieve a list of JSON files inside the subdirectory
    json_file_list = [file.path for file in os.scandir(subdirectory_path)]

    # Process each JSON file in the subdirectory
    for json_source_path in json_file_list:

        # Read and parse the source JSON file
        with open(json_source_path, 'r') as source_json_file:
            source_data = json.load(source_json_file)
            json_destination_path = json_source_path.replace("annotsleft", "annots")

            # Read and parse the existing destination JSON file
            with open(json_destination_path, 'r') as destination_json_file:
                destination_data = json.load(destination_json_file)

            # Update the destination JSON data
            destination_data["annots"][0]["handl2d"] = source_data["annots"][0]["handl2d"]
            destination_data["annots"][0]["bbox_handl2d"] = source_data["annots"][0]["bbox_handl2d"]

            # Write the modified data back to the destination JSON file
            with open(json_destination_path, 'w') as destination_json_file:
                json.dump(destination_data, destination_json_file, indent=4)

# Remove the source directory
shutil.rmtree(source_directory)
```

# Traingulation

Now that we have json file for both and right hand with keypoints in 2d we have to exploit easymocap function to triangulate them. In the config section is present a yml file regarding multiple views for 1 person. It can be modified removing parts not related to hands and then used the trinagulator function or not modify it and then use the triangulate function specific for hands. We follow first procedure.The yml file can be also modified to play with parameters that are used for the triangulation.Finally a video is created to visualize the results.

```python
import yaml

file_path = '/content/EasyMocap/config/recon/mv1p.yml'

with open(file_path, 'r') as yaml_file:
    data = yaml.safe_load(yaml_file)

del data['args']['keys']['face2d']
del data['args']['keys']['keypoints2d']

with open(file_path, 'w') as yaml_file:
    yaml.dump(data, yaml_file, default_flow_style=False)

print(f'"face2d: face2d" entry has been deleted from {file_path}')
```
