In the following code syncronization is performed beacuse the central camera has different frame rate respect to top left back and top right camera. The first frames are deleted.

```python
def syncronization(folder_path, to_delete):

  files = os.listdir(folder_path)

  files.sort()

  for i in range(to_delete):
      file_to_delete = os.path.join(folder_path, files[i])
      os.remove(file_to_delete)

  for i, file in enumerate(files[to_delete:], start=1):
      old_file_path = os.path.join(folder_path, file)
      new_file_name = f'{i:06d}.jpg'
      new_file_path = os.path.join(folder_path, new_file_name)
      os.rename(old_file_path, new_file_path)

syncronization("/content/new_data/images/01", 120)
syncronization("/content/new_data/images/02", 163)
syncronization("/content/new_data/images/03", 163)
```

```python
import json

# Define the path to your main folder
main_folder_path = '/content/new_data/annotsl'

# Get a list of sub-folders within the main folder
subfolders = [f.path for f in os.scandir(main_folder_path)]

# Loop through each sub-folder
for subfolder_path in subfolders:

    # Get a list of JSON files within the sub-folder
    json_files = [f.path for f in os.scandir(subfolder_path)]

    # Loop through the JSON files in the sub-folder
    for json_file_path in json_files:

        # You can now read and process the JSON file
        with open(json_file_path, 'r') as json_file:
            data = json.load(json_file)
            json_file_path_dest = json_file_path.replace("annotsl", "annots")

            # Step 1: Read the existing data from the destination JSON file
            with open(json_file_path_dest, 'r') as json_file:
                existing_data = json.load(json_file)

            # Step 2: Modify or add new data to the existing data
            existing_data["annots"][0]["handl2d"] =  data["annots"][0]["handl2d"]
            existing_data["annots"][0]["bbox_handl2d"] =  data["annots"][0]["bbox_handl2d"]

            # Step 3: Write the updated data back to the JSON file
            with open(json_file_path_dest, 'w') as json_file:
                json.dump(existing_data, json_file, indent=4)

#delete source folder
shutil.rmtree('/content/new_data/annotsl')
```

# Triangulation

Now that we have json file for both and right hand with keypoints in 2d we have to exploit easymocap function to triangulate them. In the config section is present a yml file regarding multiple views for 1 person. It can be modified removing parts not related to hands and then used the trinagulator function or not modify it and then use the triangulate function specific for hands. We follow first procedure.The yml file can be also modified to play with parameters that are used for the triangulation.Finally a video is created to visualize the results.

```python
import yaml

file_path = '/content/EasyMocap/config/recon/mv1p.yml'

with open(file_path, 'r') as yaml_file:
    data = yaml.safe_load(yaml_file)

del data['args']['keys']['face2d']
del data['args']['keys']['keypoints2d']

with open(file_path, 'w') as yaml_file:
    yaml.dump(data, yaml_file, default_flow_style=False)


```

```python
import yaml

# Define the path to your YAML file
yaml_file_path = '/content/EasyMocap/config/recon/triangulator-mv1p-total.yml'  # Replace with the actual file path

# Define the new value for 'dist_max' in handr2d
new_dist_max = 0.035  # Replace with your desired value

# Load the YAML file
with open(yaml_file_path, 'r') as file:
    data = yaml.safe_load(file)

# Update the 'dist_max' value in the 'handr2d' section
if 'args' in data and 'config' in data['args']:
    config = data['args']['config']
    if 'handr2d' in config:
        config['handr2d']['dist_max'] = new_dist_max

# Save the modified YAML back to the file
with open(yaml_file_path, 'w') as file:
    yaml.dump(data, file)
```
